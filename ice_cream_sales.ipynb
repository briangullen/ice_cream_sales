## Question 3: <a class="anchor" id="question3"></a>
Answer: See line and bar graphs below

```python
sns.lineplot(data=ice_cream, x='Month', y='Total Units').set_title('Total Ice Cream Sales by Month')
```

```python
x = ice_cream['Month']
vanilla = ice_cream['Vanilla']
chocolate = ice_cream['Chocolate']
strawberry = ice_cream['Strawberry']
neapolitan = ice_cream['Neapolitan']
mango = ice_cream['Mango']
lime = ice_cream['Lime']
fig, ax = plt.subplots()
ax.plot(x, vanilla, label='vanilla')
ax.plot(x, chocolate, label='chocolate')
ax.plot(x, strawberry, label='strawberry')
ax.plot(x, neapolitan, label='neapolitan')
ax.plot(x, mango, label='mango')
ax.plot(x, lime, label='lime')
ax.set_title('Ice Cream Flavors Sold by Month')
ax.set_xlabel('Month')
ax.set_ylabel('Total Units')
ax.legend()
```

```python
sns.barplot(x='Month', y='Total Profit', data=ice_cream).set_title('Total Profit per Month')
```
